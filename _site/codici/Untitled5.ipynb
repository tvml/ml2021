```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colors = ['#008fd5', '#fc4f30', '#e5ae38', '#6d904f', '#8b8b8b', '#810f7c', '#137e6d', '#be0119', '#3b638c', '#af6f09']
```

```python
x = np.linspace(-2,4,1000)
```

```python
def loss01(x):
    if x<0:
        return 1
    else:
        return 0
```

```python
def squared(x):
    return (x-1)**2
```

```python
def logistic(x):
    return np.log(1+np.exp(-x))/np.log(2)
```

```python
def exponential(x):
    return np.exp(-x)
```

```python
def hinge(x):
    return max(0,1-x)
```

```python
fig = plt.figure(figsize=(16,8))
plt.axvline(0,-1,2, color='k', linewidth=1.2)
plt.axvline(1,.18,.22, color='k', linewidth=1.2)
plt.axhline(0,-2,4, color='k', linewidth=1.2)
plt.plot(x,np.vectorize(loss01)(x), color=colors[0], linewidth=2, label='0/1')
plt.plot(x,np.vectorize(squared)(x), color=colors[1], linewidth=2,label='squared')
plt.plot(x,np.vectorize(exponential)(x), color=colors[2], linewidth=2,label='exponential')
plt.plot(x,np.vectorize(logistic)(x), color=colors[3], linewidth=2,label='logistic')
plt.plot(x,np.vectorize(hinge)(x), color=colors[6], linewidth=2,label='hinge')
plt.ylim(-1,4)
plt.xlim(-2,4)
plt.xticks([])
plt.yticks([])
sns.despine(left=True, bottom=True, right=True)
plt.legend(frameon=False)
plt.show()
```
