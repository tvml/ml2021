```python
sigma=1
s=1
```

```python
def fun(a,M):
    eps = a**M
    return(1+a)/((1-a)*(1-eps**(2)))

def fun1(a):
    return(1+a)/(1-a)

fun = np.vectorize(fun)
```

```python
def ff(a,M):
    eps = a**M
    l = (1+a)/((1-a)*(1-eps**(2)))
    first = sigma/l
    second = s*eps**2
    return first+second

ff = np.vectorize(ff)
```
